```python
skills_user = ['ms office','ms excel', 'english','adobe photoshop', 'adobe illustrator','video editing' ]
```

```python
universal_skill = pd.read_excel('skill.xlsx')
```

```python
universal_skill.sample(25)
```

```python
text = []
for i,j,k in zip(universal_skill.Category,universal_skill['Sub Category'],universal_skill['Skill Term']):
    #print(i.strip().split(),'\n',j.strip().split(),'\n',k.strip().split())
    text.extend(i.strip().split())
    text.extend(j.strip().split())
    text.extend(k.strip().replace('(','').replace(')','').split())
    print(text)

    #break
```

```python
print(text)
```

```python
def jobs_calc(job, text):
    """Counting how many occurances"""
    count = 0
    for uni_word in job:
        if uni_word in text:
            count += 1
    return count
```

```python
def accuracy(skill_list, text, job_list):
    """Calculate match accuracy for each job."""
    # skills_required = []
    # skills_match = []
    match = []  # match depending upon the skills required.
    for job in job_list:
        x = jobs_calc(job, text)
        y = jobs_calc(skill_list, job)
        # skills_required.append(x)
        # skills_match.append(y)
        match.append((y / (x+1)) * 100)
    return match
```

```python
res = accuracy(skills_user,text,jobs_desc)
```
